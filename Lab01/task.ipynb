```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('bg.png',0)
plt.subplot(121),plt.imshow(img)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img,cmap = 'gray')
plt.title('Gray Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt
from numpy import asarray

image = cv2.imread("bg.png")

plt.imshow(image)
plt.show()

data = asarray(image)
print(data)
```

```python
import imutils
from matplotlib import pyplot as plt

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()

R1 = imutils.rotate(image, 90)
plt.imshow(cv2.cvtColor(R1, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image (90 degrees)')
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()

R, G, B = cv2.split(image)
C = 255 - R
M = 255 - G
Y = 255 - B

fig, axes = plt.subplots(2, 3, figsize=(12, 8))
axes[0, 0].imshow(R, cmap='gray')
axes[0, 0].set_title("RED COMPO")
axes[0, 1].imshow(G, cmap='gray')
axes[0, 1].set_title("Green COMPO")
axes[0, 2].imshow(B, cmap='gray')
axes[0, 2].set_title("Blue COMPO")
axes[1, 0].imshow(C, cmap='gray')
axes[1, 0].set_title("Cyan COMPO")
axes[1, 1].imshow(M, cmap='gray')
axes[1, 1].set_title("Magenta COMPO")
axes[1, 2].imshow(Y, cmap='gray')
axes[1, 2].set_title("Yellow COMPO")
plt.tight_layout()
plt.show()
```
